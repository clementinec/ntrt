```python
dat = pd.read_csv('trtcleaned.csv',index_col=0)
dat = dat[dat['(hr)']>0] 
```

```python
fig = plt.figure(figsize=(10,7))
plt.plot(dat['(hr)'],dat['Supply1'],marker = '*',linestyle = '--',label='Supply Temperature ($\degree$ F)',markersize=2) 
plt.plot(dat['(hr)'],dat['Return1'],marker = '*',linestyle = '--',label='Return Temperature ($\degree$ F)',markersize=2) 
plt.plot(dat['(hr)'],dat['(Btu/hr-ft)'],marker = '*',linestyle = '--',label='Heat Injected (Btu/hr-ft)',markersize=2) 
plt.margins(x=0) 
plt.legend() 
plt.grid(True,alpha=0.2)
plt.savefig('TRTraw.png',dpi=150) 
```

```python
dat['Supply_c'] = (dat['Supply1']-32.0)/9*5
dat['Return_c'] = (dat['Return1']-32.0)/9*5
dat['HeatIn_W'] = 0.29307*dat['(Btu/hr-ft)']*1340/1000
```

```python
matplotlib.rcParams.update({'font.size': 15})

fig,ax = plt.subplots(figsize=(10,7))
plt.plot(dat['(hr)'],dat['Supply_c'],marker = '*',linestyle = '--',label='Supply Temperature (Inlet) ($\degree$C)',markersize=2,color = 'orange') 
plt.plot(dat['(hr)'],dat['Return_c'],marker = '*',linestyle = '--',label='Return Temperature (Outlet) ($\degree$C)',markersize=2,color = 'blue') 
ax2 = plt.twinx()
ax2.plot(dat['(hr)'],dat['HeatIn_W'],marker = '.',linestyle = '--',label='Heat Input (kW)',markersize=4,color = 'grey') 
ax2.set_ylabel('Heat Injected ($kW$)')
ax.set_xlabel('No. of hours since the beginning of TRT (hr)')
ax.set_ylabel('Temperature measured ($\degree C$)')
plt.margins(x=0) 
ax.legend() 
ax.grid(True,alpha=0.3)
plt.savefig('TRT_SI.png',dpi=150) 
```
